```python
#W7A1
a=list(map(int,input().split()))
k=int(input())
m,l,r=0,0,len(a)-1
while l<=r:
    m=(l+r)//2
    if a[m]==k:
        print(m)
        break
    elif a[m]>k:
        r=m-1
    else:
        l=m+1
else:
    print(-1)
```

```python
#W7A2
def fidx(a,x):
    fx=-1
    m,l,r=0,0,len(a)-1
    while l<=r:
        m=(l+r)//2
        if a[m]==x:
            fx=m
            r=m-1
        elif a[m]>x:
            r=m-1
        else:
            l=m+1
    return fx
def lidx(a,x):
    lx=-1
    m,l,r=0,0,len(a)-1
    while l<=r:
        m=(l+r)//2
        if a[m]==x:
            lx=m
            l=m+1
        elif a[m]>x:
            r=m-1
        else:
            l=m+1
    return lx       
a=list(map(int,input().split()))
x=int(input())
a.sort()
if fidx(a,x)==-1:
    print(0)
else:
    print(lidx(a,x)-fidx(a,x)+1)
```

```python
#W7A4
a=list(map(int,input().split()))
d={}
for i in range(len(a)):
    if a[i] in d:
        d[a[i]]+=1
    else:
        d[a[i]]=1
m,p=0,a[0]
for i in range(len(a)-1,-1,-1):
    if d[a[i]]>=m:
        m=d[a[i]]
        p=a[i]
print(p,"xuat hien nhieu nhat, som nhat,",d[p],"lan")
```

```python
#W7A5
a=list(map(int,input().split()))
x=int(input())
r=0
for i in range(len(a)-1):
    for j in range(i+1,len(a)):
        if a[i]+a[j]==x:
            r+=1
print(r)
```
