```python
#W4A2
import math
n=int(input())
while n<0:
    n=int(input())
if n < 2:
    print("Not prime")
else:
    l = int(math.sqrt(n)) + 1
    for i in range(2, l):
        if n % i == 0:
            print("Not prime")
            break
    else:
        print("Prime")
```

```python
#W4A6
m=[1]*int(1e5)
m[0]=m[1]=0
for i in range(2,int(1e5)):
    if m[i]==1:
        for j in range(i*2,int(1e5),i):
            m[j]=0
a,b=map(int,input().split())
s=0
for i in range(a,b+1):
    if m[i]==1:
        s+=i
print(s)
```
