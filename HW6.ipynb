```python
#W6A4
str=input().split()
r={}
for i in str:
    k,v=i.split(':')
    if k not in r:
        r[k]=[v]
    else:
        r[k].append(v)
print(r)
```

```python
#W6A5
a=list(map(int,input().split()))
r={'positives':0, 'negatives':0, 'zeros':0}
for i in a:
    if i>0:
        r['positives']+=1
    elif i<0:
        r['negatives']+=1
    else:
        r['zeros']+=1

print(r)
```

```python
#W6A9
d1=eval(input())
d2=eval(input())
for k,v in d2.items():
    if k in d1:
        d1[k]+=v
    else: 
        d1[k]=v
s=sorted(d1)
r={}
for i in s:
    r[k]=d1[k]
print(r)
```

```python
#W6A10
a=list(map(int,input().split()))
k=int(input())
t=set()
for i in range(len(a)-1):
    for j in range(i+1,len(a)):
        if a[i]+a[j]==k:
            t.add((a[i],a[j]))
t=sorted(t)
print(t)
```

```python
#W6A12
a=list(map(int,input().split()))
d={}
for i in a:
    if i not in d:
        d[i]=1
    else:
        d[i]+=1
m=0
id=float('inf')
for i in d:
    if d[i]>=m:
        m=d[i]
        id=i
    elif d[i]==m:
        if i<id:
            id=i
print(id)
```

```python
#W6A13
a=input().split()
d={}
for i in range(0,len(a),2):
    k=a[i]
    v=a[i+1]
    d[k]=v
r={}
for k,v in d.items():
    r[v]=k
print(r)
```

```python
#W6A15
a=input().split()
b=int(input())
d={}
r={}
for i in range(0,len(a),2):
    k=a[i]
    v=int(a[i+1])
    d[k]=v
for i in d:
    if d[i]>b:
        r[i]=d[i]
print(r)
```
