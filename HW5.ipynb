```python
#W5A1
def maxx(a,b):
    if a>b:
        return a
    return b

a,b=map(int,input().split())
print(maxx(a,b))
```

```python
#W5A2
def swapp(a,b)->tuple[int,int]:
    t=a
    a=b
    b=t
    return a,b 
a,b=map(int,input().split())
print(swapp(a,b))
```

```python
#W5A3
def snt(n):
    if n<2:
        return 0
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            return 0
    return 1
n=int(input())
print(snt(n))
```

```python
#W5A4
def shh(n):
    s=0
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            s+=i
    return s==n
n=int(input())
print(shh(n))
```

```python
#W5A5
def tim(k,a):
    for i in range(len(a)):
        if a[i]==k:
            return i
    return -1
a=list(map(int,input().split()))    
k=int(input())
print(tim(k,a))
```

```python
#W5A6
def gt(n):
    if n==0 or n==1:
        return 1
    return n*gt(n-1)
n=int(input())
print(gt(n))
```

```python
#W5A7
def cal(a,b,o):
    if o=='+':
        return a+b
    elif o=='-':
        return a-b
    elif o=='*':
        return a*b
    elif o=='/':
        if b==0:
            return "Error"
        return a/b
    else:
        return "Invalid operator"
a,b=map(int,input().split())
o=input()
print(cal(a,b,o))
```

```python
#W5A9
def tong(n):
    if n==0:
        return 0
    return n%10 + tong(n//10)
n=int(input())
print(tong(n))
```

```python
#W6A19
def dm(a):
    m=min(a)
    for i in range(len(a)):
        if a[i]==m:
            return i
a=list(map(int,input().split()))
print(dm(a))
```

```python
#W6A20
def ff(a,k):
    for i in range(len(a)):
        if a[i]==k:
            return i
    return -1
a=list(map(int,input().split()))
k=int(input())
print(ff(a,k))
```

```python
#W6A21
s=input()
w=s.split()
r={}
for i,c in enumerate(w):
    if c in r:
        r[c].append(i)
    else:
        r[c]=[i]
for k in r:
    r[k]=tuple(r[k])
print(r)
```
