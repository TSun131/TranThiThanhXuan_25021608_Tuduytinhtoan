```python
#Bài 3
import math

a, b, c = map(float, input("Nhap a b c: ").split())

if a <= 0 or b <= 0 or c <= 0 or a + b <= c or a + c <= b or b + c <= a:
    print("Khong la ba canh cua mot tam giac")
else:
    if a == b == c:
        print("Tam giac deu")
    elif a == b or b == c or a == c:
        print("Tam giac can")
    else:
        sides = sorted([a, b, c])
        if sides[0]**2 + sides[1]**2 == sides[2]**2:
            print("Tam giac vuong")
        else:
            print("Tam giac thuong")
    P = a + b + c
    p = P / 2
    S = math.sqrt(p * (p - a) * (p - b) * (p - c))
    print(f"Chu vi: {P:.2f}")
    print(f"Dien tich: {S:.2f}")
```

```python
#Bài 5
a,b,c=map(int,input().split())
D=b**2-4*a*c
if D<0:
    print("Phuong trinh vo nghiem")
elif D==0:
    x=-b/(2*a)
    print("Phuong trinh co nghiem kep:",x)
else:
    x1=(-b+math.sqrt(D))/(2*a)
    x2=(-b-math.sqrt(D))/(2*a)
    print("Phuong trinh co hai nghiem phan biet:",x1,x2)
```

```python
#Bài 8
a,b,c,d,m,n=map(int,input().split())
D=(a*d-b*c)
Dx=(m*d-b*n)    
Dy=(a*n-c*m)
if D==0:
    if Dx==0 and Dy==0:
        print("Hệ phương trình vô số nghiệm")
    else:
        print("Hệ phương trình vô nghiệm")
else:
    x = Dx / D
    y = Dy / D
    print(f"Hệ phương trình có nghiệm duy nhất: x = {x}, y = {y}")
```
