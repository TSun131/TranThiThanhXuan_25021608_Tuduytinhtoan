```python
#W7A16
a=list(map(int,input().split()))
d={}
b,c=1,0
for i in a:
    if i not in d:
        d[i]=1
    else:
        d[i]+=1
for i in d:
    if d[i]!=i:
        b=0
        if d[i]<i:
            c+=d[i]
        else:
            c+=d[i]-i
if b:
    print("YES")
else:
    print("NO", c)
```

```python
#W7A20
def scs(n):
    r=0
    while n:
        r+=1
        n//=10
    return r

a=list(map(int,input().split()))
k=int(input())
s=0
for i in a:
    if scs(i)==k:
        s+=i
print(s)
```
